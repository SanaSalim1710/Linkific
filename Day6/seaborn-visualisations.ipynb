```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
titanic = sns.load_dataset("titanic")
```

```python
iris_df = sns.load_dataset("iris")
```

### Pair Plot

```python
sns.pairplot(iris_df)
```

```python
sns.pairplot(titanic,hue="sex")
```

```python
tips_df = sns.load_dataset("tips")
sns.pairplot(tips_df, hue='sex')
```

### Box Plot

```python
sns.boxplot(tips_df,x="day", y="total_bill", hue="sex")
```

```python
sns.boxplot(titanic,x="class",y="age",hue="sex")
```

### Violin Plot

```python
sns.violinplot(tips_df,x="day",y="total_bill",hue="sex")
```

```python
sns.violinplot(titanic,x="survived",y="age",hue="sex")
```

### Bar Plot

```python
sns.barplot(titanic, x="sex", y="survived")
```

```python
sns.barplot(titanic, x="class", y="survived")
```

### Count Plot

```python
sns.countplot(titanic, x="sex")
```

```python
sns.countplot(titanic,x="survived")
```

### Strip Plot

```python
plt.figure(figsize=(10,8))
sns.stripplot(titanic, x="class",y="age")
```

```python
plt.figure(figsize=(10,8))
sns.stripplot(titanic,x="sex",y='age')
```

### Line Plot

```python
sns.lineplot(titanic, x="age",y="fare")
```

### Histogram

```python
sns.histplot(iris_df, x="sepal_length")
```

```python
sns.histplot(titanic,x="age",kde=True)
```

### Heatmap

```python
titanic_matrix = titanic.corr(numeric_only=True)
titanic_matrix
```

```python
sns.heatmap(titanic_matrix,annot=True,cmap="Blues")
```

```python
iris_matrix = iris_df.corr(numeric_only=True)
iris_matrix
sns.heatmap(iris_matrix,annot=True,cmap="Blues")
```

### Styling

```python
sns.set_style("white")
plt.figure(figsize=(8,5))
sns.jointplot(iris_df,x="sepal_length",y="petal_length",kind="reg")
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,5))
sns.set_context('paper',font_scale=1.4)
sns.jointplot(iris_df,x="sepal_length",y="petal_length",kind="reg")
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,5))
sns.set_context('talk',font_scale=1.4)
sns.jointplot(iris_df,x="sepal_length",y="petal_length",kind="reg")
```

```python
sns.set_style("ticks")
plt.figure(figsize=(8,5))
sns.set_context('paper',font_scale=0.8)
sns.jointplot(iris_df,x="sepal_length",y="petal_length",kind="reg")
sns.despine(left=True)
```

```python
plt.figure(figsize=(10,8))
sns.stripplot(titanic,x="class",y='age',palette="afmhot")
```
