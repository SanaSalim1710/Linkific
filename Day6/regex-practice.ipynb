# Regex

```python
import re
```

### Extract emails from text

```python
text = "lalala hbdhd 121 j322 sana_salim@gmail.com regular expressions 12345 ssal123@yahoo.com ssal123@student.monash.edu"
emails = re.findall(r"[a-zA-Z0-9_.]+@[a-zA-Z]+\.[a-zA-Z.]+",text)
for item in emails:
    print(item)
```

### Extract phone numbers from text

```python
text = "01113225888 sjnwjenw222 22 +6010209820 (+44)2021234 +69 113355667 303-339-4455 1234567891011"
numbers = re.findall(r"[(]?[+]?\d{1,3}[-\.\s)]?\d{3}[-\.\s]?\d{3,}",text)
for item in numbers:
    print(item)
```

### Extract dates in different formats

```python
text = "lala 12/2/2024 jh2jh 13/01/22 kw9j 2024-02-2 okok 2-3-22"
dates = re.findall(r"\d{1,4}[/-]\d{1,2}[/-]\d{1,4}",text)
for item in dates:
    print(item)
```

### Validate Input Formats

Validate email

```python
email = "example101@gmail.com"
fake_email = "la&!101/@gma11.com"
valid1 = bool(re.match(r"[a-zA-Z0-9_.]+@[a-zA-Z]+\.[a-zA-Z.]+",email))
valid2 = bool(re.match(r"[a-zA-Z0-9_.]+@[a-zA-Z]+\.[a-zA-Z.]+",fake_email))
if valid1:
    print(email, "is a valid email")
else:
    print(email, "is not a valid email")
if valid2:
    print(fake_email, "is a valid email")
else:
    print(fake_email, "is not a valid email")
```

Validate password

```python
passwords = ["haha1234","Haha1234!","Aa!","haha1234!","password!"]
for password in passwords:
    valid = bool(re.match(r"^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&])[a-zA-Z0-9!@#$%^&*()?]{8,}$",password))
    if valid:
        print(password, "is a valid password")
    else:
        print(password, "is not a valid password")

```

### Clean messy text

```python
text = "  This is @ corn] 211   ! owi! "
# remove trailing and leaading whitespace
text = text.strip()
print(text)
# remove extra spaces
# replace one or more whitespaces with single whitespace 
clean = re.sub(r"\s+"," ",text)
# only alphanumeric characters (remove special characters)
clean = re.findall(r"\w+",text)
clean =' '.join(clean)
print(clean)
```
