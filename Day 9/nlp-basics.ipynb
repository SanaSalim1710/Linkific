# Text Cleaning

```python
def clean_text(text):
    # make everything lowercase
    text = text.lower()
    # remove trailing and leading whitespace
    text = text.strip()
    # remove extra whitespace
    text = re.sub(r"\s+"," ",text)
    # remove html tags
    text = re.sub("<.*?>", "", text) 
    # remove special characters
    text = re.sub("[^a-zA-Z]", " ", text)     
    return text

texts = ["A wonderful little production. <br /><br />",
"    The filming technique is very unassuming-very old-time-BBC fashion and gives a comforting, and sometimes discomforting, sense of realism",
"to the entire piece. <br /><br />The actors are extremely well chosen- Michael Sheen not only 'has got all the polari' but ",
"he has all the voices down pat too! You can truly see      the seamless editing guided by the references to Williams' diary entries, "]
for text in texts:
    print(clean_text(text))
```

### Remove Stopwords

Filter out common words

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
text = "This sentence has stopwords that will be removed."
sw = set(stopwords.words('english'))
tokens = word_tokenize(text.lower())
removed = []
for word in tokens:
    if word not in sw:
        removed.append(word)
print(removed)
text2 = "This is tangerine sticky sweet but or honeybee of expectations"
tokens2 = word_tokenize(text2.lower())
removed2 = []
for word in tokens2:
    if word not in sw:
        removed2.append(word)
print(removed2)
```

### Stemming

Reduce words to root or base form (remove prefixes and suffixes)

SnowballStemmer: Aggressive and might lead to overstemming

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer(language='english')
text = ['dancing', 'singing', 'crashed', 'steadily', 'cases']
stemmed = []
for word in text:
    stemmed.append(stemmer.stem(word))
print("Original words:", text)
print("Stemmed words:", stemmed)
```

LancasterStemmer: Aggressive and might lead to overstemming. Less efficient than snowball for large datasets.

```python
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
text = ['dancing', 'singing', 'crashed', 'steadily', 'cases']
stemmed = []
for word in text:
    stemmed.append(stemmer.stem(word))
print("Original words:", text)
print("Stemmed words:", stemmed)
```

PorterStemmer: Output may not always be a real word. English is the only language supported.

```python
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()
text = ['dancing', 'singing', 'crashed', 'steadily', 'cases']
stemmed = []
for word in text:
    stemmed.append(stemmer.stem(word))
print(text)
print("Stemmed:", stemmed)
```

### Lemmatization

Converts words into base or dictionary words called lemma. Only uses real words, unlike stemming.

```python
nltk.download('wordnet') 
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
text = "The foxes jumped quickly over the fence. The dogs are running. Children are screaming."
tokens = word_tokenize(text)
lemmatized = []
for word in tokens:
    lemmatized.append(lemmatizer.lemmatize(word))
print(text)
print(f"Lemmatized: {lemmatized}")
# need pos tagging for better results
```

```python
lemmatized = []
for word in tokens:
    lemmatized.append(lemmatizer.lemmatize(word,pos="v"))
print(text)
print(f"Lemmatized: {lemmatized}")
```

```python
from nltk import pos_tag
pos_tags = pos_tag(tokens)
for word in pos_tags:
    print(word)
```

### Bag of words

Convert text into a collection of words and count frequency of words.

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
sentences = ["I love dogs and dogs love me","NLP is great", "NLP sucks"]
bow_vectorizer = CountVectorizer()
bow_matrix = bow_vectorizer.fit_transform(sentences)
vocab = bow_vectorizer.get_feature_names_out()
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=vocab)
bow_df.index = [f'Doc {i+1}' for i in range(len(sentences))]
count = 1
for sentence in sentences:
    print(f"Doc {count}:", sentence)
    count += 1
print(vocab,"\n")
print(bow_df.to_string())
```

### TF-IDF

### Term Frequency-Inverse Document Frequency
It measures how often a term appears in a document (term frequency).
It reduces the weight of common words and increases the rate of rare words (inverse document frequency).

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)  # reuse corpus from BoW section
vocab_tfidf = tfidf_vectorizer.get_feature_names_out()
index = []
for i in range(len(sentences)):
    index.append(f"Doc {i+1}")
tfidf_df = pd.DataFrame(tfidf_matrix.toarray().round(3),columns=vocab_tfidf,index=index)
print(tfidf_df.to_string())
```
