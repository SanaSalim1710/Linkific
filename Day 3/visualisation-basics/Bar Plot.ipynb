```python
import matplotlib.pyplot as plt
import numpy as np

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
ariana_expenses = [180, 130, 150, 200, 250, 190, 120]
olivia_expenses = [170, 150, 190, 120, 150, 180, 200]
w = 0.4
x = np.arange(len(days))

plt.title("Daily Expenditure")
plt.ylabel("Day")
plt.xlabel("Expenses")

plt.bar(x-w/2, ariana_expenses, w, label = "Ariana")
plt.bar(x+w/2, olivia_expenses, w, label = "Olivia")
plt.xticks(x, days)

plt.legend()
plt.show()
```
