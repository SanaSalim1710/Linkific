```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
ariana_expenses = [120, 130, 150, 180, 190, 200, 250]
olivia_expenses = [110, 120, 140, 190, 200, 180, 220]

plt.title("Daily Expenditure")
plt.ylabel("Day")
plt.xlabel("Expenses")
plt.scatter(days, ariana_expenses, color='blue',label="Ariana")
plt.scatter(days, olivia_expenses, color='red',label="Olivia")

plt.legend()
plt.show()
```
