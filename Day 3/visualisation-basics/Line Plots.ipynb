```python
import matplotlib.pyplot as plt

plt.title("Weekly Temperatures in Various Cities")
plt.xlabel("Day")
plt.ylabel("Temperature")

kannur = [22, 26, 27, 28, 31, 27, 33]
delhi = [30, 29, 29, 27, 27, 26, 25]
mumbai = [24, 25, 26, 27, 28, 29, 30]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

plt. plot(days, kannur,label = "Kannur")
plt.plot(days, delhi, label = "Delhi")
plt.plot(days, mumbai, label = "Mumbai")
plt.legend()
plt.show()
```
