### Group By Operations

Find average age of survivors

```python
avg_age = df.groupby("Survived")["Age"].mean()
print(avg_age)
```

Find number of men and women who survived

```python
survivors = df.groupby("Sex")["Survived"].sum()
print(survivors)
```

Find number of survivors in each cabin

```python
cabin_survivors = df.groupby("Cabin")["Survived"].sum()
print(cabin_survivors)
cabin_survivors.max()
```

```python
# total number of survivors
print(df[df["Survived"]==1])
```

Find average age by passenger class

```python
avg_age_class = df.groupby("Pclass")["Age"].mean()
print(avg_age_class)
```

Find port survival rate

```python
port_survival = df.groupby("Embarked")["Survived"].sum()
port_survival_rate = df.groupby("Embarked")["Survived"].sum()
survived_port = df.groupby("Embarked").size()
print(survived_port)
print(port_survival)
print(port_survival_rate)
print(len(df))
```

### Merging and joining

```python
# creating new data frame with same passenger id and new column "Child"
df_id = df[["PassengerId"]].copy()
df_id["Child"] = df["Age"] < 18

print(df_id.loc[171:184])
```

```python
df_with_child = pd.merge(df,df_id,on="PassengerId")
df_with_child.head()
```

```python
df_family = df[["PassengerId"]].copy()
df_family["Family Size"] = df["SibSp"] + df["Parch"] + 1
print(df_family["Family Size"].max())
print(df_family.loc[180])
df_family.tail()
```

```python
df_with_family = df.join(df_family["Family Size"])
df_with_family.head()
```

### Pivot Table

Average age of males and females in each passenger class

```python
pd.pivot_table(df,index="Pclass",values="Age",columns="Sex",aggfunc="mean")
```

Counts number of male and female passengers in each class

```python
df.pivot_table(values="PassengerId",index="Pclass",columns="Sex",aggfunc="count")
```
