# Analyse Titanic Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/sanasalim/Downloads/Day 4/Titanic-Dataset.csv")
```

### Are most of the survivors men or women?

```python
survived_sex = df.groupby("Sex")["Survived"].sum()
print(survived_sex)
```

```python
sex_survival_rate = df.groupby("Sex")["Survived"].mean()*100
print(sex_survival_rate)
```

```python
ax = sex_survival_rate.round(1).plot(kind="bar",title="Survival by Sex")
ax.set_ylabel("Survival Rate (%)")
for container in ax.containers:
    ax.bar_label(container)
```

### Average age by passenger class

```python
avg_age_class = df.groupby("Pclass")["Age"].mean()
print(avg_age_class)
```

```python
ax1 = avg_age_class.round(1).plot.barh()
ax1.set_xlabel("Average Age")
ax1.set_ylabel("Passenger Class")
for container in ax1.containers:
    ax1.bar_label(container)
```

### Survival rate by embarkation port

```python
port_survival = df.groupby("Embarked")["Survived"].sum()
port_survival_rate = df.groupby("Embarked")["Survived"].mean().round(4)*100
survived_port = df.groupby("Embarked").size()
print(survived_port)
print(port_survival)
print("\nPort Survival Rate:\n",port_survival_rate)
print(len(df))
```

```python
ax3 = port_survival_rate.round(1).plot.barh()
ax3.set_xlabel("Survival Rate (%)")
ax3.set_ylabel("Embarkation Port")
for container in ax3.containers:
    ax3.bar_label(container)
```
