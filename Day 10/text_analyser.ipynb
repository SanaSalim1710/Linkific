```python
import transformers
from transformers import pipeline
```

```python
class TextAnalysis:

    def __init__(self):
        self.sentiment_analysis = pipeline("sentiment-analysis", model="siebert/sentiment-roberta-large-english")
        self.ner = pipeline("ner", model= "Jean-Baptiste/roberta-large-ner-english", aggregation_strategy="simple")
        self.summariser = pipeline("summarization", model="facebook/bart-large-cnn")
    
    def analysis(self, text):
        print("Text Analysis")
        print("-----------------------------------------------------------")
        #print("Original Text:",text,"\n")

        print("Sentiment Analysis")
        # got an error abt maximum sequence length for the model being 512
        sentiment = self.sentiment_analysis(text[:512])
        print("Sentiment:", sentiment[0]['label'])
        print("Confidence Score:", sentiment[0]['score'],"\n")

        print("Named Entity Recognition")
        names = self.ner(text)
        for entity in names:
            if entity['entity_group'] == "PER":
                word = "Person"
            elif entity['entity_group'] == "LOC":
                word = "Location"
            elif entity['entity_group'] == "ORG":
                word = "Organisation"
            elif entity['entity_group'] == "MISC":
                word = "Miscellaneous"
            print(entity['word'],":", word)
            print("Confidence Score:", entity['score'])
        #print(names)
        #print(f"Total entities found: {len(names)}")
        print("\n")

        print("Text Summary")
        summary = self.summariser(text,max_length=100, min_length=30)
        print("Summary:", summary[0]['summary_text'])
```
